```python
silver_df = spark.table("silver_crypto_market")

display(silver_df)
```

```python
from pyspark.sql.functions import desc

gold_marketcap = silver_df.orderBy(
    desc("market_cap")
).limit(10)

display(gold_marketcap)
```

```python
gold_marketcap.write.mode("overwrite").saveAsTable("gold_top_marketcap")
```

```python
gold_volume = silver_df.orderBy(
    desc("total_volume")
).limit(10)

display(gold_volume)
```

```python
gold_volume.write.mode("overwrite").saveAsTable("gold_top_volume")
```

```python
gold_gainers = silver_df.orderBy(
    desc("price_change_percentage_24h")
).limit(10)

display(gold_gainers)
```

```python
gold_gainers.write.mode("overwrite").saveAsTable("gold_price_gainers")
```

```python
from pyspark.sql.functions import sum, avg, max

summary = silver_df.select(
    sum("market_cap").alias("Total Market Cap"),
    avg("current_price").alias("Average Price"),
    max("current_price").alias("Highest Price")
)

display(summary)
```

```python
from pyspark.sql.functions import sum, avg, max

summary = silver_df.select(
    sum("market_cap").alias("total_market_cap"),
    avg("current_price").alias("average_price"),
    max("current_price").alias("highest_price")
)

display(summary)
```

```python
summary.write.mode("overwrite").saveAsTable("gold_market_summary")
```

```python
from pyspark.sql.functions import col

dim_coin = silver_df.select(
    col("id"),
    col("symbol"),
    col("name")
).dropDuplicates()

display(dim_coin)
```

```python
dim_coin.write.mode("overwrite").saveAsTable("dim_coin")
```

```python
from pyspark.sql.functions import to_date

dim_date = silver_df.select(
    to_date("last_updated").alias("date")
).dropDuplicates()

display(dim_date)
```

```python
dim_date.write.mode("overwrite").saveAsTable("dim_date")
```

```python
fact_crypto = silver_df.select(
    "id",
    "current_price",
    "market_cap",
    "market_cap_rank",
    "total_volume",
    "high_24h",
    "low_24h",
    "price_change_percentage_24h",
    "circulating_supply",
    "last_updated",
    "ingestion_time"
)

display(fact_crypto)
```

```python
fact_crypto.write.mode("overwrite").saveAsTable("fact_crypto_market")
```
