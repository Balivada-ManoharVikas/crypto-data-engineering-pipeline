```python
import requests

url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 20,
    "page": 1,
    "sparkline": "false"
}

response = requests.get(url, params=params)

print("Status Code:", response.status_code)
```

```python
data = response.json()

print(data[0])
```

```python
import pandas as pd

pdf = pd.DataFrame(data)

pdf.head()
```

```python
required_columns = [
    "id",
    "symbol",
    "name",
    "current_price",
    "market_cap",
    "market_cap_rank",
    "total_volume",
    "high_24h",
    "low_24h",
    "price_change_percentage_24h",
    "circulating_supply",
    "last_updated"
]

pdf = pdf[required_columns]

display(pdf)
```

```python
numeric_columns = [
    "current_price",
    "market_cap",
    "market_cap_rank",
    "total_volume",
    "high_24h",
    "low_24h",
    "price_change_percentage_24h",
    "circulating_supply"
]

for col in numeric_columns:
    pdf[col] = pd.to_numeric(pdf[col], errors="coerce")
```

```python
spark_df = spark.createDataFrame(pdf)

display(spark_df)
```

```python
spark_df.write.mode("overwrite").saveAsTable("bronze_crypto_market")
```

```python
bronze_df = spark.table("bronze_crypto_market")

display(bronze_df)
```

```python
from pyspark.sql.functions import current_timestamp

silver_df = bronze_df.withColumn(
    "ingestion_time",
    current_timestamp()
)

display(silver_df)
```

```python
silver_df = silver_df.dropDuplicates(["id"])

display(silver_df)
```
